```python
#1. Positives and Negatives
#Variable Declaration
pos_counter = 0
neg_counter = 0
zero_counter = 0

#Input loop
while True:
    user_input = input("Input: ")
    if user_input != "a" :
        if user_input[len(user_input)-1].isnumeric():
            user_input = int(user_input)
            if user_input > 0:
                pos_counter += 1
            elif user_input < 0:
                neg_counter += 1
            elif user_input == 0:
                zero_counter += 1
        else:
            print("Invalid input. Try Again.")
            continue
    elif user_input == "a":
        break

#Output
print()
print("Positive - {}".format(pos_counter))
print("Negative - {}".format(neg_counter))
print("Zeroes - {}".format(zero_counter))
```

```python
#2: The Four-Six Pour Over Coffee Brewing Method 

#Amount of Coffee
while True:
    amount_coffee = input("Enter amount of coffee in grams: ")
    if amount_coffee.isnumeric():
        amount_coffee = int(amount_coffee)
        break
    else:
        print("You entered an invalid amount. Please try again.")
        continue

#Water Amount and Part Division
total_water = amount_coffee*15
first_part = total_water*0.4
second_part = total_water*0.6

#Sweet or Acidic
while True:
    brew_taste = input("Do you want your coffee sweet or acidic? (type sweet or acidic): ")
    if brew_taste == "sweet":
        first_pour = first_part*0.4
        second_pour = first_part*0.6
        break
    elif brew_taste == "acidic":
        first_pour = first_part*0.6
        second_pour = first_part*0.4
        break
    else:
        print("You entered an invalid option. Please try again.")
        continue
        
#Strong or Not
while True:
    strength = input("Do you want your coffee strong (Y or N)").lower()
    if strength == "y":
        break
    elif strength == "n":
        break
    else:
        print("You entered an invalid option. Please try again.")
        continue

#Output
print("Water pour plan:")
print("First Part ({})g:".format(round(first_part,1)))
print("Pour #1: {} grams".format(round(first_pour,1)))
print("Pour #1: {} grams".format(round(second_pour,1)))
print("Second Part ({})g:".format(round(second_part,1)))
if strength == "y":
    division = 0.25
    remaining_pour = round(second_part*division,1)
    print("Pour #3: {} grams".format(remaining_pour))
    print("Pour #4: {} grams".format(remaining_pour))
    print("Pour #5: {} grams".format(remaining_pour))
    print("Pour #6: {} grams".format(remaining_pour))
elif strength == "n":
    division = 0.5
    remaining_pour = second_part*division
    print("Pour #3: {} grams".format(remaining_pour))
    print("Pour #4: {} grams".format(remaining_pour))
print("Total water: {}g".format(total_water))
```

```python
#3: Restaurant Cashier Simulation
def cashier():
    print("Welcome! Here is our menu:")
    print("\n\t- Chicken Burger: Php 60")
    print("\t- Sisig Burrito: Php 70")
    print("\t- Beef Kebab Bowl: Php 80")
    print("\t- Calamansi Juice: Php 20")
    chicken_burger_val = 60
    sisig_burrito_val = 70
    beef_kebab_val = 80
    cal_juice_val = 20
    order_val = 0
    order_count = 0
    while order_count < 5:
        print()
        order = input("What will you have? ")
        print('("chicken burger", "sisig burrito", "beef kebab bowl", "calamansi juice", or "/checkout" to checkout now)')
        print(order)
        if order == "chicken burger":
            order_val += chicken_burger_val
            order_count += 1
        elif order == "sisig burrito":
            order_val += sisig_burrito_val
            order_count += 1
        elif order == "beef kebab bowl":
            order_val += beef_kebab_val
            order_count += 1
        elif order == "calamansi juice":
            order_val += cal_juice_val
            order_count += 1
        elif order == "/checkout":
            break
        else:
            print("Ooops, that's not on our menu!")
            continue
        print("That's Php {} so far.".format(order_val))
    print("\nThat will be a total of Php {}".format(order_val))
    payment = input("How much cash will you pay? ")
    payment = int(payment)
    if payment == order_val:
        print("Thank you for giving the exact amount. Enjoy your order!")
    elif payment > order_val:
        change = payment - order_val
        print("Thank you for your payment. Your change is Php {}. Enjoy your order!".format(change))
    elif payment < order_val:
        print("Insufficient payment. Transaction could not be completed.")
    
cashier()
```

```python
#4:Hollow squares with Diagonal Shades
def square(N):
    N = int(N)
    n = N
    N_total = (N*2)-1
    n_total = N_total
    N_inverse = N_total-N
    n_inverse = N_inverse
    for i in range(0,N_total): #vertical
        for j in range(0,n_total): #horizontal
            if i == 0 or i == N_total-1:
                print("*", end = " ")
            elif i in range(0,N) and j in range(0,n): 
                print("*", end = " ")
            elif j == 0 or j == n_total-1:
                print("*", end = " ")
            elif i in range(N_inverse,N_total) and j in range(n_inverse,n_total):
                print("*", end = " ")
            else:
                print(" ", end = " ")
        print()
        
square(input("Input: "))
```
